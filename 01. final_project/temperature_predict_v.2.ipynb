Разработать модель машинного обучения для прогнозирования конечной температуры стали в металлургическом процессе с целью оптимизации энергопотребления. **Задачи проекта:**

1. **Загрузка и первичное исследование данных**

3. **Объединение данных и создание финального датасета**
   - Создание единой таблицы с агрегированными признаками для каждой партии
   - Проверка на наличие асинхронности данных временных меток
   - Корреляционный анализ и отбор наиболее значимых признаков

     - Случайный лес
     - Градиентный бустинг (XGBoost, LightGBM)
     - Нейронные сети (PyTorch)
   - Проведение кросс-валидации и подбор гиперпараметров с использованием Optuna
   - Оценка качества моделей по метрике MAE

                                                                                                                
6. **Тестирование и валидация лучшей модели**

   - Анализ важности признаков и интерпретируемости модели

```python
def get_steel_tables_dimensions(db_path):
    """Получить размерность таблиц"""
    with sqlite3.connect(db_path) as conn:
        # Получаем список всех таблиц
        all_tables = pd.read_sql_query(
            "SELECT name FROM sqlite_master WHERE type='table'", conn
        )
        
        # Фильтруем только steel-таблицы (начинающиеся с data_)
        steel_tables = [t for t in all_tables['name'] if t.startswith('data_')]
        
        dimensions = []
        
        for table_name in steel_tables:
            # Получаем количество строк
            count_query = f"SELECT COUNT(*) as row_count FROM {table_name}"
            row_count = pd.read_sql_query(count_query, conn)['row_count'].iloc[0]
            
            # Получаем количество столбцов
            pragma_query = f"PRAGMA table_info({table_name})"
            columns_info = pd.read_sql_query(pragma_query, conn)
            col_count = len(columns_info)
            
            dimensions.append({
                'table': table_name,
                'rows': row_count,
                'columns': col_count,
                'dimension': f"({row_count},{col_count})"
            })
        
        # Сортируем по количеству строк (по убыванию)
        dimensions_df = pd.DataFrame(dimensions)
        return dimensions_df
```

```python
def query_table_name(table_name):
    table = (
        f"""
        SELECT *
        FROM {table_name}
        """
    )
    return pd.read_sql_query(table, conn)
```

Аналогично как и графиком выше

Рассчитаем коэффициент мощности по формуле:
$$
\cos\varphi = \frac{P}{\sqrt{P^2 + Q^2}}
$$

где:
- $Q$ - реактивная мощность (вар)
- $P$ - активная мощность (вар)

```python
df_arc['Коэффициент_мощности'] = df_arc['Активная мощность'] / np.sqrt(
    df_arc['Активная мощность']**2 + df_arc['Реактивная мощность']**2
)
```

```python
search_plot(df_arc['Коэффициент_мощности'],'Распределение коэффициента мощности',30)
```

- Коэффициент мощности находится в диапазоне 0,7-0,88.

```python
df_arc['Эффективность_нагрева'] = (
    df_arc['Активная мощность'] / 
    (df_arc['Активная мощность'] + abs(df_arc['Реактивная мощность']))
)
```

### Предобработка df_temp

Так как заказчик сказал, что сплав всегда нагревался, то это необходимо проверить

```python
df_temp['Время замера'] = pd.to_datetime(df_temp['Время замера'])
df_temp['Температура'] = pd.to_numeric(df_temp['Температура'], errors='coerce')
```

```python
df_sorted = df_temp.sort_values(['key', 'Время замера'])
```

```python
def get_valid_first_last(group):
    """Возвращает первый и последний замеры только если оба валидны"""
    # Проверяем, что первый и последний замеры не NaN
    first_row = group.iloc[0]
    last_row = group.iloc[-1]
    
    if pd.isna(first_row['Температура']) or pd.isna(last_row['Температура']):
        return None  # Пропускаем партию с NaN в первом или последнем замере
    
    return pd.Series({
        'Время_начала': first_row['Время замера'],
        'Начальная_температура': first_row['Температура'],
        'Время_окончания': last_row['Время замера'],
        'Конечная_температура': last_row['Температура']
    })
```

```python
df_temp_agg = df_sorted.groupby('key').apply(get_valid_first_last).dropna().reset_index()
```

```python
df_temp_agg = df_temp_agg[
    (df_temp_agg['Начальная_температура']>1500) & (df_temp_agg['Конечная_температура']>1500)
]
```

```python
def objective_rf(trial):
    
    
    params = {
        'model__n_estimators': trial.suggest_int('n_estimators', 50, 300),
        'model__max_depth': trial.suggest_int('max_depth', 3, 20),
        'model__min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'model__min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
        'model__max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None]),
        'model__bootstrap': trial.suggest_categorical('bootstrap', [True, False])
    }
    
    # Обновляем параметры в pipeline
    rf_pipeline.set_params(**params)
    
    # Кросс-валидация
    scores = cross_val_score(
        rf_pipeline, X_train, y_train, 
        cv=kf, scoring='neg_mean_absolute_error', n_jobs=-1
    )
    
    return -scores.mean()  

# Оптимизация
study_rf = optuna.create_study(
    direction='minimize',
    sampler=TPESampler(seed=RANDOM_STATE)
)
study_rf.optimize(objective_rf, n_trials=30)

print("Best Random Forest params:")
print(study_rf.best_params)
print(f"Best CV MAE: {study_rf.best_value:.4f}")
```

### Нейронная сеть

```python
scaler = StandardScaler()
X_train_nn = scaler.fit_transform(X_train)
X_test_nn = scaler.transform(X_test)
```

```python
X_train_tensor = torch.FloatTensor(X_train_nn)
y_train_tensor = torch.FloatTensor(y_train.values)
X_test_tensor = torch.FloatTensor(X_test_nn)
y_test_tensor = torch.FloatTensor(y_test.values)

# Создаем DataLoader
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print(f"Размерность данных: {X_train_tensor.shape[1]} признаков")
```

```python
hidden_size1=128
hidden_size2=64
hidden_size3=32
```

```python
class TemperaturePredictor(nn.Module):
    def __init__(self, input_size, hidden_size1=128, hidden_size2=64, hidden_size3=32, dropout_rate=0.2):
        super(TemperaturePredictor, self).__init__()
        
        self.network = nn.Sequential(
            nn.Linear(input_size, hidden_size1),
            nn.BatchNorm1d(hidden_size1),
            nn.ReLU(),
            #nn.Dropout(dropout_rate),
            
            nn.Linear(hidden_size1, hidden_size2),
            nn.BatchNorm1d(hidden_size2),
            nn.ReLU(),
            #nn.Dropout(dropout_rate),
            
            nn.Linear(hidden_size2, hidden_size3),
            nn.BatchNorm1d(hidden_size3),
            nn.ReLU(),
            #nn.Dropout(dropout_rate),
            
            nn.Linear(hidden_size3, 1)
        )
    
    def forward(self, x):
        return self.network(x).squeeze()
```

```python
input_size = X_train_tensor.shape[1]
model = TemperaturePredictor(input_size)
```

```python
def train_model(model, train_loader, test_loader, learning_rate=0.001, epochs=100, patience=10):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    
    criterion = nn.L1Loss()  
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-5)
    
    train_losses = []
    
    for epoch in range(epochs):
        # Training
        model.train()
        train_loss = 0
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            
            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item() * X_batch.size(0)
        
        train_loss /= len(train_loader.dataset)
        train_losses.append(train_loss)
        
        if (epoch + 1) % 100 == 0:
            print(f'Epoch {epoch+1}/{epochs}, Train MAE: {train_loss:.4f}')
    
    return model, train_losses, train_losses[-1]
```

```python
def objective_nn(trial):
    """Objective function for Neural Network optimization"""
    
    params = {
        'hidden_size1': trial.suggest_int('hidden_size1', 64, 256),
        'hidden_size2': trial.suggest_int('hidden_size2', 32, 128),
        'hidden_size3': trial.suggest_int('hidden_size3', 16, 64),
        'dropout_rate': trial.suggest_float('dropout_rate', 0.1, 0.5),
        'learning_rate': trial.suggest_float('learning_rate', 1e-4, 1e-2, log=True),
        'batch_size': trial.suggest_categorical('batch_size', [32, 64, 128])
    }
    
    
    train_loader_opt = DataLoader(train_dataset, batch_size=params['batch_size'], shuffle=True)
    test_loader_opt = DataLoader(test_dataset, batch_size=params['batch_size'], shuffle=False)
    
    # Создаем и обучаем модель
    model = TemperaturePredictor(
        input_size, 
        params['hidden_size1'], 
        params['hidden_size2'], 
        params['hidden_size3'],
        params['dropout_rate']
    )
    
    _, _, best_mae = train_model(
        model, train_loader_opt, test_loader_opt,
        learning_rate=params['learning_rate'],
        epochs=300,
        patience=15
    )
    
    return best_mae
```

```python
study_nn = optuna.create_study(direction='minimize', sampler=TPESampler(seed=RANDOM_STATE))
study_nn.optimize(objective_nn, n_trials=10)

print("Best Neural Network params:")
print(study_nn.best_params)
print(f"Best CV MAE: {study_nn.best_value:.4f}")
```

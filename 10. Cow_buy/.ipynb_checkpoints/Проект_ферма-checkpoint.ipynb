```python
# –í–≤–µ–¥–µ–º –§—É–Ω–∫—Ü–∏—é –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (10,5), legend = False)               

        );
        plt.title('–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è, %' + "\n", fontsize=22)
        plt.xlabel('–ü—Ä–æ—Ü–µ–Ω—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤', fontsize=18)
        plt.ylabel('–°—Ç–æ–ª–±—Ü—ã —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏',fontsize=18)
        plt.xlim(0, 100)  
    except:
        print('–ø—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ—Ç –∏–ª–∏ –ø—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞ ')
```

```python
#–§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ–± –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
def ShowInfo(df):
    print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –≤—ã–±–æ—Ä–∫–∏ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç',df.shape[0],'\n')
    display(df.info())
    display(df.sample(5))
    display(df.describe().T)
    print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç',df.duplicated().sum())
    pass_value_barh(df)
    
```

–ü—Ä–∏ –∏–∑—É—á–µ–Ω–∏–∏ –¥–∞–Ω–Ω—ã—Ö –±—ã–ª –∑–∞–º–µ—á–µ–Ω –≤—ã–±—Ä–æ—Å —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º 45616. –ò–∑–±–∞–≤–∏–º—Å—è –æ—Ç –Ω–µ–≥–æ –ø–µ—Ä–µ–¥ –Ω–∞—á–∞–ª–æ–º –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞.

```python
#–°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
def SearchPlot(df_col, title_name):
    df_col.plot(
    kind = 'hist',
    figsize = (8,6),
    bins = 20,
    edgecolor='black'
)
    plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ, ' + title_name)
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
    plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, ' + title_name)
    plt.axvline(df_col.mean(), color='red', label='—Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ')
    plt.axvline(df_col.median(), color='purple', label='–º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ')
    plt.legend(loc='upper right',fontsize='x-small')
    plt.show()

    df_col.plot(
        kind = 'box',
        figsize = (8,6)
    )
    plt.title('–Ø—â–∏–∫ —Å —É—Å–∞–º–∏, ' + title_name)
    plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏—è, '+ title_name)
    plt.show()

    display(df_col.describe().T)
```

```python
def SearchPlotCategory(df,col,title_name):
    
    ratio = round(df.groupby(col)[col].count()/df[col].count()*100)
    
    ratio.plot(
    kind = 'pie',
    autopct='%1.1f%%',
    labels = ratio.index,
    figsize=(6,6),
    title = '–ö—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö '+ title_name,
    ylabel = '',
    fontsize = 12
    )
    display(ratio)
```

---

### –¢—Ä–µ—Ç—å—è –º–æ–¥–µ–ª—å


#### –í—ã–±–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –£—Å—Ç—Ä–∞–Ω–µ–Ω–∏–µ –Ω–µ–ª–∏–Ω–µ–π–Ω–æ—Å—Ç–∏.

–î–æ–±–∞–≤–∏–º –∫ –ø—Ä–µ–¥—ã–¥—É—â–µ–π –≤—ã–±–æ—Ä–∫–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫ "–ò–º—è –æ—Ç—Ü–∞"

```python
ml_3 = df.drop(['id', 'milk_protein_ratio','milk_taste','fat'], axis=1)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –°–ü–û –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫
ml_3['spo_category'] = ml_3['sugar_protein_ratio'].apply(lambda x: 1 if x > 0.91 else 0)
display(ml_3['spo_category'].value_counts())

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –ø—Ä–∏–∑–Ω–∞–∫ –≠–ö–ï 
ml_3['eke_2'] = ml_3['eke'] ** 0.5

# –£–±–∏—Ä–∞–µ–º –Ω–µ–≤–∞–∂–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã
ml_3 = ml_3.drop(['sugar_protein_ratio', 'eke'], axis=1)
ml_3.head(1)
```

#### –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–∫–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é

```python
X = ml_3.drop('income_milk_kg',axis=1)
y = ml_3['income_milk_kg']
X_train, X_test, y_train, y_testt = train_test_split(X,y, random_state=RANDOM_STATE)
```

#### –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –ú–û

```python
#–°–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ —Å —Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫.
numeric_col = ['raw_protein_g','eke_2']
category_col =['breed','type_landscape','breed_dad','age','dad_name','spo_category']

#–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è
encoder = OneHotEncoder(drop='first', sparse=False)
X_train_ohe = encoder.fit_transform(X_train[category_col])
X_test_ohe = encoder.transform(X_test[category_col])
encoder_col_names = encoder.get_feature_names()
```

```python
#–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º —á–∏—Å–ª–æ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[numeric_col])
X_test_scaled = scaler.transform(X_test[numeric_col])
```

```python
# –°–æ–±–∏—Ä–∞–µ–º –≤—ã–±–æ—Ä–∫–∏ 
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=numeric_col)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=numeric_col)

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
```

#### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

```python
ml_study_3 = LinearRegression()

ml_study_3.fit(X_train, y_train)

predictions = ml_study_3.predict(X_test)

r2_3 = r2_score(y_test, predictions)
mae_3 = mean_absolute_error(y_test, predictions)
rmse_3 = mean_squared_error(y_test, predictions, squared=False)
print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–µ—Ç–µ—Ä–º–µ–Ω–∞—Ü–∏–∏:',r2_3,'\n'
      '–°—Ä–µ–¥–Ω—è—è –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –æ—à–∏–±–∫–∞:',mae_3,'\n'
      '–°—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞:',rmse_3
     )
```

#### –ê–Ω–∞–ª–∏–∑ –æ—Å—Ç–∞—Ç–∫–æ–≤

```python
#–ü–æ—Å—á–∏—Ç–∞–µ–º –æ—Å—Ç–∞—Ç–∫–∏
residuals_3 = y_test - predictions
```

```python
# –°—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 8))
axes[0].hist(residuals_3, bins=40, color='c', edgecolor='black')
zero = axes[0].vlines(x=0, colors='m', ymin=0, ymax=16, linestyles='--', label='–Ω—É–ª–µ–≤–∞—è –æ—Ç–º–µ—Ç–∫–∞')
median = axes[0].vlines(x=residuals_3.mean(), colors='red', ymin=0, ymax=16, label='—Å–º–µ—â–µ–Ω–∏–µ ')
axes[0].legend()
axes[0].set_title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Å—Ç–∞—Ç–∫–æ–≤')
axes[0].set_xlabel('–û—Å—Ç–∞—Ç–∫–∏')

# –°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–∏—Å–ø–µ—Ä—Å–∏–∏
axes[1].scatter(predictions, residuals_3, alpha=0.7)
axes[1].set_xlabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏')
axes[1].set_ylabel('–û—Å—Ç–∞—Ç–∫–∏')
axes[1].set_title('–ê–Ω–∞–ª–∏–∑ –¥–∏—Å–ø–µ—Ä—Å–∏–∏')
zero = axes[1].hlines(y=0,xmin=5000,xmax=6750,color='r')
plt.show()
```

- __–£—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç—å:__ –¥–∏—Å–ø–µ—Ä—Å–∏—è –æ—Å—Ç–∞—Ç–∫–æ–≤ —Å—Ç–∞–ª–∞ –ø–ª–æ—Ç–Ω–µ–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–π –º–æ–¥–µ–ª–∏. –î–∏—Å–ø–µ—Ä—Å–∏—è —Å—Ç–∞–ª–∞ –ª—É—á—à–µ.

```python
# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ r2_score
print('---------C—Ä–∞–≤–Ω–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –¥–µ—Ç–µ—Ä–º–µ–Ω–∞—Ü–∏–∏---------\n')
scores_compared = {'–ø–µ—Ä–≤–æ–π': r2_1, '–≤—Ç–æ—Ä–æ–π': r2_2, '—Ç—Ä–µ—Ç—å–µ–π': r2_3}
for k, v in scores_compared.items():
    print(f'R2_score {k} –º–æ–¥–µ–ª–∏: {v}')

print('\n---------C—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤---------\n')

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤
residuals_compared = {'–ø–µ—Ä–≤–æ–π': residuals_1, '–≤—Ç–æ—Ä–æ–π': residuals_2, '—Ç—Ä–µ—Ç—å–µ–π': residuals_3}
for k, v in residuals_compared.items():
    print(f'–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤ {k} –º–æ–¥–µ–ª–∏: {v.mean()}')
```

#### –í—ã–±–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

–î–ª—è –æ—Ü–µ–Ω–∫–∏ –≤–∫—É—Å–∞ –º–æ–ª–æ–∫–∞ –ø–æ–¥–æ–π–¥—É—Ç —Ç–∞–∫–∏–µ –≤—Ö–æ–¥–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∫–∞–∫ –∂–∏—Ä–Ω–æ—Å—Ç—å, –±–µ–ª–æ–∫, —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –∫–æ—Ä–æ–≤—ã –∏ —Ç–∏–ø–∞ –ø–∞—Å—Ç–±–∏—â–∞, –∞ —Ç–∞–∫–∂–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –µ–µ –∫–æ—Ä–º–∞. –ú–æ–∂–Ω–æ —Å–º–µ–ª–æ —É–±—Ä–∞—Ç—å —É–¥–æ–π –∏ –≤—ã—Å–æ–∫–æ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—â–∏–µ –º–µ–∂–¥—É —Å–æ–±–æ–π –≤—Ö–æ–¥–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Ç–∞–∫–∏–µ –∫–∞–∫ `–ü–æ—Ä–æ–¥–∞ –æ—Ç—Ü–∞` –∏ `–ò–º—è –æ—Ç—Ü–∞`, –æ—Å—Ç–∞–≤–∏–º –≤—Ç–æ—Ä–æ–µ.


```python
ml_log = df.copy(deep=True)
ml_log['milk_taste']=ml_log['milk_taste'].apply(lambda x: 1 if x=='–≤–∫—É—Å–Ω–æ' else 0)

```

```python
ml_log.head(1)
```

```python
ml_log = ml_log.drop(['id', 'income_milk_kg','breed_dad'], axis=1)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –°–ü–û –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫
ml_log['spo_category'] = ml_log['sugar_protein_ratio'].apply(lambda x: 1 if x > 0.91 else 0)
display(ml_log['spo_category'].value_counts())

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –ø—Ä–∏–∑–Ω–∞–∫ –≠–ö–ï 
ml_log['eke_2'] = ml_log['eke'] ** 0.5

# –£–±–∏—Ä–∞–µ–º –Ω–µ–≤–∞–∂–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã
ml_log = ml_log.drop(['sugar_protein_ratio', 'eke'], axis=1)
ml_log.head(1)
```

<div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>
 
üëç –ê–≥–∞, –∫–æ–¥–∏—Ä—É–µ–º —Ç–∞—Ä–≥–µ—Ç.



```python
# –í—ã–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ, –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ
X = ml_log.drop('milk_taste', axis=1)
y = ml_log['milk_taste']

X_train, X_valid, y_train, y_valid = train_test_split(
    X,
    y,
    random_state=RANDOM_STATE,
    stratify=y
)

cat_col_names = ['breed', 'type_landscape', 'age','dad_name','spo_category']
num_col_names = ['raw_protein_g', 'fat', 'milk_protein_ratio','eke_2']

encoder = OneHotEncoder(drop='first', sparse=False)
X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
X_valid_ohe = encoder.transform(X_valid[cat_col_names])
encoder_col_names = encoder.get_feature_names()

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[num_col_names])
X_valid_scaled = scaler.transform(X_valid[num_col_names])

X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_valid_ohe = pd.DataFrame(X_valid_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_valid_scaled = pd.DataFrame(X_valid_scaled, columns=num_col_names)

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_valid = pd.concat([X_valid_ohe, X_valid_scaled], axis=1)
```

<div class="alert alert-block alert-danger">
<h2><s> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"></h2>
  
<b>–ù–∞ –¥–æ—Ä–∞–±–æ—Ç–∫—Éü§î:</b>
    
–ó–¥–µ—Å—å —Ç–∞–∫–∂–µ –Ω—É–∂–Ω–æ –±—É–¥–µ—Ç –¥–æ–±–∞–≤–∏—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–∫–∞–∫ –º–∏–Ω–∏–º—É–º, –∏–º—è –ø–∞–ø—ã –±—ã–∫–∞ –¥–æ–±–∞–≤–ª—è–µ–º, –∞ –≤—ã—Å–æ–∫–æ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—â–∏–π —Å –Ω–∏–º –ø—Ä–∏–∑–Ω–∞–∫ –ø–æ—Ä–æ–¥—ã –ø–∞–ø—ã_–±—ã–∫–∞ –º–æ–∂–Ω–æ —É–±—Ä–∞—Ç—å).
    
–ü—Ä–∏–≥–æ–¥–∏—Ç—Å—è –≤—Å–µ, –∫—Ä–æ–º–µ id –∏ —É–¥–æ—è.

<div class="alert alert-block alert-info">
    –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–æ


```python
clf = LogisticRegression(random_state=RANDOM_STATE)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_valid)

clf_accuracy = accuracy_score(y_pred, y_valid)
clf_recall = recall_score(y_pred, y_valid)
clf_precision = precision_score(y_pred, y_valid)

print(f'Accuracy: {clf_accuracy}')
print(f'Recall: {clf_recall}')
print(f'Precision: {clf_precision}')
```

–ù–∞–¥–µ—é—Å—å, —à—É—Ç–∫–∞. –í—ã—à–µ –≤–µ—Ä–Ω–æ –ø–∏—à–µ—à—å, —á—Ç–æ precision.

### –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ø–æ—Ä–æ–≥–∞

```python
# –í—ã–≤–µ–¥–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∞ predict_proba() –∏ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –æ—Ç–≤–µ—Ç—ã
y_proba = clf.predict_proba(X_valid)[:, 1]

proba_data = pd.DataFrame(zip(y_valid, y_proba),
                          columns=['y_valid', 'y_proba']).sort_values(by='y_proba', ascending=False)

proba_data.head()
```

```python
thresholds = [round(i,2) for i in np.linspace(0.5,1,num = 10,endpoint=False)]
print(thresholds) 
```

<div class="alert alert-warning">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>
    
<b>–ù–µ–∫–æ—Ç–æ—Ä—ã–µ –∑–∞–º–µ—á–∞–Ω–∏—è –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏üí°:</b> 

–û–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ, –≤—ã—à–µ —É –Ω–∞—Å –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –±—ã–ª –ø–æ—Ä–æ–≥ –≤ 0.5 –∏ –æ–Ω –Ω–µ –ø–æ–¥–æ—à–µ–ª –ø–æ precision. –ó–Ω–∞—á–∏—Ç –∑–¥–µ—Å—å –ø–æ—Ä–æ–≥ –Ω–∏–∂–µ 0.5 –Ω–∞–º —Ç–æ—á–Ω–æ –Ω–µ –Ω—É–∂–µ–Ω –±—É–¥–µ—Ç. –ú–æ–∂–Ω–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –æ—Ç 0.6 –¥–æ 0.9
        
</div> 



```python
columns=[]
for i in thresholds:
	columns.append('y_pred_'+str(i))
	proba_data['y_pred_'+str(i)] = proba_data['y_proba'].apply(lambda x: 1 if x>=i else 0)

# –≤—ã–≤–µ–¥–µ–º 5 —Å–ª—É—á–∞–π–Ω—ã—Ö —Å—Ç—Ä–æ–∫
display(proba_data.sample(5))
```

```python
# –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å—á–µ—Ç–∞ –º–µ—Ç—Ä–∏–∫ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ—Ä–æ–≥–∞
def ThresholdChoice(threshold_list, proba_df):
    result_df = pd.DataFrame(columns=['–ø–æ—Ä–æ–≥', 'precision', 'recall', '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ_FP', '—á–∏—Å–ª–æ_–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö_–æ–±—ä–µ–∫—Ç–æ–≤',
                                      '–≤—Å–µ–≥–æ_–æ–±—ä–µ–∫—Ç–æ–≤_–∫–ª–∞—Å—Å–∞_1'])

    for threshold in threshold_list:
        column_name = 'y_pred_' + str(threshold)

        # Precision
        precision = proba_df.loc[proba_df[column_name] == 1, 'y_valid'].mean()

        # –ß–∏—Å–ª–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤
        predicted_objects = proba_df[column_name].sum()

        # –í—Å–µ–≥–æ –æ–±—ä–µ–∫—Ç–æ–≤ –∫–ª–∞—Å—Å–∞ 1
        total_objects_class_1 = proba_df['y_valid'].sum()

        # Recall
        recall = proba_df.loc[(proba_df[column_name] == 1) & (proba_df['y_valid'] == 1)].shape[0] / total_objects_class_1

        # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–æ–∂–Ω–æ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (FP)
        false_positives = proba_df.loc[(proba_df[column_name] == 1) & (proba_df['y_valid'] == 0)].shape[0]

        # –î–æ–±–∞–≤–ª—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
        result_df = result_df.append({
            '–ø–æ—Ä–æ–≥': threshold,
            'precision': precision,
            'recall': recall,
            '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ_FP': false_positives,
            '—á–∏—Å–ª–æ_–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö_–æ–±—ä–µ–∫—Ç–æ–≤': predicted_objects,
            '–≤—Å–µ–≥–æ_–æ–±—ä–µ–∫—Ç–æ–≤_–∫–ª–∞—Å—Å–∞_1': total_objects_class_1,
        }, ignore_index=True)

    return result_df
```

```python
ThresholdChoice(thresholds,proba_data)
```

–ü—Ä–∏ —ç—Ç–æ–º, –µ—Å–ª–∏ –≤ –∏—Ç–æ–≥–µ –Ω–∏ –æ–¥–Ω–∞ –∏–∑ –∫–æ—Ä–æ–≤ –Ω–µ –ø–æ–¥–æ–π–¥–µ—Ç –ø–æ –≤–∫—É—Å—É ‚àí —ç—Ç–æ –Ω–æ—Ä–º–∞–ª—å–Ω—ã–π –æ—Ç–≤–µ—Ç, —Ç–∞–∫–æ–π –æ—Ç–≤–µ—Ç –≤–ø–æ–ª–Ω–µ –º–æ–∂–µ—Ç –ø–æ–ª—É—á–∞—Ç—å—Å—è –∏ –≤ —Ä–µ–∞–ª—å–Ω—ã—Ö –ø—Ä–æ–µ–∫—Ç–∞—Ö. –í–∞–∂–Ω–æ –±—É–¥–µ—Ç –ø—Ä–æ–≥–æ–≤–æ—Ä–∏—Ç—å, –ø–æ—á–µ–º—É —Ç–∞–∫ –≤—ã—à–ª–æ, –∫–∞–∫–∏–µ –ø—Ä–æ–±–ª–µ–º—ã –µ—Å—Ç—å —Å –º–æ–¥–µ–ª—å—é –∏ –¥–∞–Ω–Ω—ã–º–∏.
